```
Given the head of a linked list, reverse the nodes of the list k at a time, and return the modified list.

k is a positive integer and is less than or equal to the length of the linked list. If the number of nodes is not a multiple of k then left-out nodes, in the end, should remain as it is.

You may not alter the values in the list's nodes, only nodes themselves may be changed.

Example 1:
Input: head = [1,2,3,4,5], k = 2
Output: [2,1,4,3,5]

Example 2:
Input: head = [1,2,3,4,5], k = 3
Output: [3,2,1,4,5]

Constraints:
The number of nodes in the list is n.
1 <= k <= n <= 5000
0 <= Node.val <= 1000
```

## Approach 1
- Make a helper function to reverse k nodes at a time and give start , end and (k+1)th node
- Call this helper function repeteadly for every kth node until all the list is reversed
- After each call link the previous reversed end node to newly reversed start node

**T(n)=O(N*K)**

**S(n)=O(1)**

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:  
    def reverse_k_nodes(self,head,k):
        if not head:
            return None,None,None
        elif not head.next:
            return head,None,None

        #check the size of list is smaller or not 
        curr=head
        c=0
        while curr and c<k:
            curr=curr.next
            c+=1
        if c!=k: #skipping the smaller size list(last list)
            return head,None,None

        #reverse the list by k nodes
        curr=prev=head
        ahead=head.next 
        k-=1
        while ahead and k>0:       
            curr=ahead
            ahead=ahead.next
            curr.next=prev
            prev=curr
            k-=1

        #return the start,end and next node[(K+1)th node] in new linked list
        return curr,head,ahead


    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        start,end,next_list_node=None,None,None
        result=prev_end=ListNode(0)

        while True:
            #reverse the list k item             
            start,end,next_list_node=self.reverse_k_nodes(head,k)
            
            #attach the previously reversed list end to this reversed start
            prev_end.next=start

            #if end is not present means reverse list is not required anymore
            if not end:
                break
            
            #updating the previously reversed linked list and the head of linked list ot be reversed
            prev_end=end
            head=next_list_node

        return result.next
````
