```
You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Example 1:
Input: l1 = [2,4,3], l2 = [5,6,4]
Output: [7,0,8]
Explanation: 342 + 465 = 807.

Example 2:
Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]
Output: [8,9,9,9,0,0,0,1]
 
Constraints:
The number of nodes in each linked list is in the range [1, 100].
0 <= Node.val <= 9
It is guaranteed that the list represents a number that does not have leading zeros.
```

## Approach 1
- Create a new node which store the sum of given list
- Maintain a carry variable to store the current carry of the sum

**T(n)=O(n)**

**S(n)=O(1)**

````python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        
        carry=0
        head=ListNode()
        curr_node=head
    
        
        while l1 and l2:
            #compute the value to be added 
            val=(l1.val+l2.val+carry)%10
            #carry value which is propogated to next addition
            carry=(l1.val+l2.val+carry)//10
            
            #linking new node to the resultant linked-list
            new_node=ListNode(val)
            curr_node.next=new_node
            curr_node=new_node
            
            #increment the node pointers
            l1=l1.next
            l2=l2.next
        
        while l1:
            #compute the value to be added 
            val=(l1.val+carry)%10
            #carry value which is propogated to next addition
            carry=(l1.val+carry)//10
            
            #linking new node to the resultant linked-list
            new_node=ListNode(val)
            curr_node.next=new_node
            curr_node=new_node
            
            #increment the node pointers
            l1=l1.next
            
            
        while l2:
            #compute the value to be added 
            val=(l2.val+carry)%10
            #carry value which is propogated to next addition
            carry=(l2.val+carry)//10
            
            #linking new node to the resultant linked-list
            new_node=ListNode(val)
            curr_node.next=new_node
            curr_node=new_node
            
            #increment the node pointers
            l2=l2.next   
        
        #the addition require extra node 
        if carry > 0:
            #linking new node to the resultant linked-list
            new_node=ListNode(carry)
            curr_node.next=new_node
            curr_node=new_node
            
        return head.next
````
