```
Given a binary tree, find the lowest common ancestor (LCA) of two given nodes in the tree.

According to the definition of LCA on Wikipedia: “The lowest common ancestor is defined between two nodes p and q as the lowest node in T that has both p and q as descendants (where we allow a node to be a descendant of itself).”

Example 1:
Input: root = [3,5,1,6,2,0,8,null,null,7,4], p = 5, q = 1
Output: 3
Explanation: The LCA of nodes 5 and 1 is 3.
                                                                                                    
Example 2:
Input: root = [3,5,1,6,2,0,8,null,null,7,4], p = 5, q = 4
Output: 5
Explanation: The LCA of nodes 5 and 4 is 5, since a node can be a descendant of itself according to the LCA definition.
 
Constraints:
The number of nodes in the tree is in the range [2, 105].
-109 <= Node.val <= 109
All Node.val are unique.
p != q
p and q will exist in the tree.
```

## Approach 1
- Recursive solution 
    
**T(n)=O(n)**

**S(n)=O(n)**

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        paths=[]
        #find the path of a given value in tree
        def find_node_path(root,val,path):
            nonlocal paths
            if not root:
                return
            #adding the element to the list
            path.append(root)
            
            if root.val==val:
                #found the node adding the path to paths
                paths.append(path.copy())
                return
            
            if root.left:
                find_node_path(root.left,val,path)
                #removing the last occurance of the item
                path.remove(root.left)
                
            if root.right:
                find_node_path(root.right,val,path)
                #removing the last occurance of the item
                path.remove(root.right)

        find_node_path(root,p.val,[])
        find_node_path(root,q.val,[])
        
        path1=paths[0]
        path2=paths[1]
        
        #find 1st nonmacthing node
        for i in range(min(len(path1),len(path2))):
            if path1[i].val!=path2[i].val:
                return path1[i-1]
        #if either of list exuasted one p/q is LCA
        return path1[min(len(path1),len(path2))-1]
````

## Approach 2
- Recursive solution 
    
**T(n)=O(n)**

**S(n)=O(logn)**

````python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        #if no nodes present in tree
        if not root:
            return None
        #if p or q is root node
        if root.val == p.val or root.val == q.val:
            return root
        
        #postorder traversal of tree to check node is found in tree or not
        left=self.lowestCommonAncestor(root.left,p,q) #Left
        right=self.lowestCommonAncestor(root.right,p,q) #Right
        #if both left and right are not None means p&q present in different subtree of root
        #hence root become LCA
        if left and right: #Root
            return root
            
        #if left is None then LCA is in right
        if right:
            return right
        else:
            return left
````
